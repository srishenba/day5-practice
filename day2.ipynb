```python
n=int(input("Enter number of terms for Fibonacci sequence:"))

a, b = 0, 1
if n <= 0:
    print("Please enter a positive integer.")
elif n == 1:
    print("Fibonacci sequence up to", n, "term:")
    print(a)
else:
    print("Fibonacci sequence up to", n, "terms:")
    print(a)
    print(b)
    for _ in range(2, n):
        next_fib = a + b
        print(next_fib)
        a = b
        b = next_fib
```
